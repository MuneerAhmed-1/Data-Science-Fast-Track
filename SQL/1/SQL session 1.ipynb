# Basic Query Structure

```python
def read_q(q):
    return pd.read_sql_query(q, conn)
```

```python
q = 'select * from sqlite_master'
read_q(q)
```

# Where
- Where in urdu -> jhn pr yay hu
- rows ko filter krny k liyay huta hy

```python
read_q('SELECT ShareWomen, Major FROM recent_grads WHERE ShareWomen > 0.5')
```

`Use ''' ''' or """ """ for multiline query`

```python
read_q(''' SELECT ShareWomen, Major 
        FROM recent_grads 
        WHERE ShareWomen > 0.5
        LIMIT 10''')
```

# Operators
`<, >, <=, >=, <>, =, AND, OR, NOT`

```python
read_q(''' SELECT ShareWomen, Major 
        FROM recent_grads 
        WHERE ShareWomen > 0.5 AND ShareWomen < 0.8''')
```

# GROUP BY
- HAVING is always use in GROUPBY
- But we use WHERE and HAVING in GROUPBY

```python
read_q("""SELECT DISTINCT Major_category, avg(ShareWomen)
FROM recent_grads 
GROUP BY Major_Category""")
```

```python
read_q("""SELECT DISTINCT Major_category, avg(ShareWomen)
FROM recent_grads 
GROUP BY 1""")
```

# HAVING
- `HAVING is only put in these type of functions like: avg, sum, count `

```python
read_q("""SELECT DISTINCT Major_category, avg(ShareWomen) AVG_SW
FROM recent_grads 
GROUP BY 1
HAVING AVG_SW > 0.5""")
```

```python
read_q("""SELECT DISTINCT Major_category, avg(ShareWomen) AVG_SW
FROM recent_grads 
WHERE Major_category != 'Social Science'
GROUP BY 1
HAVING AVG_SW > 0.5""")
```
