# JOINS
`using a common column (which is present in both tables), you can join two or more tables to retrieve information.`

Return Pakistan and its cities

```python
# country name, city name
```

```python
read_q(""" select f.name, c.name
from facts f inner join 
cities c on 
f.id = c.facts_id
""")
```

```python
read_q(""" select f.name, c.name
from facts f inner join 
cities c on 
f.id = c.facts_id
where f.id = 132
""")
```

# Task 2
- return evry country and its capital city

```python
read_q(""" select f.name Country, c.name Cities
from facts f inner join
cities c on
f.id = c.facts_id
where c.capital = 1 and f.id = 132
""")
```

# Task 3
- return the name of the country and its relative city and its population which has a population of more than 10 million.

```python
read_q(""" select f.name Country, c.name City, c.population City_population
from facts f inner join
cities c on 
f.id = c.facts_id
where c.population > 10000000
""")
```

# Task 4
- Country name, Total population, Urban Areas Population

```python
read_q(""" select f.name Country, f.population Total_Population, sum(c.population) Urban_population
from facts f inner join cities c
on f.id = c.facts_id
group by f.name
order by 3 desc
""")
```

# Task 5
- find out the percentage of urban to total population

```python
read_q(""" select f.name Country, f.population Total_Population, sum(c.population) Urban_population,
round((sum(c.population)/cast(f.population as float)) * 100, 3) per_Urban
from facts f inner join cities c
on f.id = c.facts_id
group by f.name
order by 3 desc
""")
```
